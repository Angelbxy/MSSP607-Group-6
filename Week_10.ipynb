```python
import pandas as pd
df1 = pd.read_csv("demographics.csv")
# Display the first few rows of the DataFrame
df1.head()
```

```python
import pandas as pd

# Load the data from the CSV file into a DataFrame
df2 = pd.read_csv("Connections.csv")

# Display the first few rows of the DataFrame
df2.head()
```

```python

import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
graph = nx.Graph()

# Add nodes with demographic attributes
for index, row in df1.iterrows():
  graph.add_node(row['Student ID'], Age=row['Age'], Gender=row['Gender'], Major=row['Major'])

# Add edges between students based on connection data
for index, row in df2.iterrows():
  graph.add_edge(row['Student ID 1'], row['Student ID 2'])


# Draw the graph
nx.draw(graph, with_labels=True, node_size=500, node_color="skyblue", font_size=8)
plt.show()
```
